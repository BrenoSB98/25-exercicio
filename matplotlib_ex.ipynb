1 - Plote a função y = 5x + 1. Onde x vai de 0 até 10. Use marcadores do tipo quadrado. Informe o título do gráfico e  eixos x e y. De as cores  vermelho aos marcadores e azul a reta.

```python
x = np.arange(11)
y = 5 * x + 1

plt.title("Função Y = 5 * x + 1")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.plot(x, y, c="b", mfc="r", marker="o") 
plt.show()
```

4 - Com base nos dados do exercício anterior. Plote dois gráficos separados (um no lado do outro). <br/> 
Em um você ira plotar o gráfico da variação dólar X real. E no outro ira plotar o PIB per Capita do Brasil, nos mesmo anos. <br/> 
Escolha o tipo de gráfico que preferir. Dados do PIB (dados fictícios): 2016: 8710,50 USD; 2017: 9928,50; 2018: 9151,58; 2019: 8897,29; 2020: 6795,32; 2021: 7500,21; 2022: 7542,34

5 - Crie um gráfico de pizza para representar os seguintes dados: <br/>
Destino de exportações do Brasil: China: 30% Estados Unidos: 40% Europa: 20% Outros: 10%

```python
paises = np.array(["China", "Estados Unidos", "Europa", "Outros"])
export_percent = np.array([30, 40, 20, 10])
cores_paises = np.array(['red', 'blue', 'green', 'orange'])

text_props = {
    'color': 'black',
    'size': 15,
    'family': 'arial'
}

plt.figure(figsize=(7,7))
plt.pie(export_percent, labels=paises, colors=cores_paises, radius=0.8, startangle=90, autopct = lambda value : '%.2f %s' % (value,'%'), 
        textprops = text_props)
plt.legend(paises, loc='lower right')
plt.tight_layout()
plt.show()
```

6 - Crie um gráfico de dispersão mostrando a relação entre peso e altura de algumas pessoas. Analise os resultados. <br/>
Dados altura = [1.60, 1.62, 1.65, 1.65, 1.70, 1.70, 1.75, 1.80. 1.85, 1.90, 1.90, 1.95, 2.00] <br/> 
Dados peso = [60, 61, 64, 67, 70, 73, 75, 80. 85, 90, 85, 95, 200]

```python
altura = np.array([1.60, 1.62, 1.65, 1.65, 1.70, 1.70, 1.75, 1.80, 1.85, 1.90, 1.90, 1.95, 2.00])
peso = np.array([60, 61, 64, 67, 70, 73, 75, 80, 85, 90, 85, 95, 100])


plt.scatter(altura, peso, color="orange")
plt.title("Relação Peso vs Altura")
plt.xlabel("Altura")
plt.ylabel("Peso")
plt.show()
```
